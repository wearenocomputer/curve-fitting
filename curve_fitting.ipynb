```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

%matplotlib inline

def normalizeSet(set):
     return np.interp(set,[np.amin(set),np.amax(set)],[-1,1])

#lets try to do curve fitting with a simple parabola
# y= ax**
def simpleParabola(x, a):
    return a*(x*x)

xvalues = normalizeSet(np.array([-3,-2,-1,0,1,2,3]))
yvalues = normalizeSet(simpleParabola(xvalues,1))

plt.plot(xvalues, yvalues)
plt.ylim([-2,2])
plt.xlim([-2,2])
plt.show()

popt, pcov = curve_fit(simpleParabola, xvalues, yvalues,p0=(1))
p1 = popt[0]

residuals = yvalues - simpleParabola(xvalues,p1)
fres = sum(residuals**2)

print (fres)
```

```python
#lets try to do curve fitting with a simple parabola but with data with an offset

xvalues = normalizeSet(np.array([-3,-2.9,-1,0,1.07,2.4,3]))
yvalues = normalizeSet(np.array([9,3.88,1.8,0,1,4,9]))


popt, pcov = curve_fit(simpleParabola, xvalues, yvalues,p0=(1))
p1 = popt[0]
residuals = yvalues - simpleParabola(xvalues,p1)
fres = sum(residuals**2)

print (fres)


newxvalues = normalizeSet(np.array([-3,-2.9,-1,0,1.07,2.4,3]))
newyvalues = normalizeSet(simpleParabola(newxvalues,3.18942604811))


plt.plot(xvalues, yvalues)
plt.plot(newxvalues, newyvalues)
plt.ylim([-2,2])
plt.xlim([-2,2])
plt.show()
```
